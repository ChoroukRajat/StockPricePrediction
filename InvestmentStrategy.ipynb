# STOCK PRICE PREDICTION MODEL COMPARISON: LSTM vs ARIMA vs RANDOM FOREST


```python
"""Stock Price Prediction: Comparing LSTM, XGBoost and Random Forest"""

# ====================== 1. IMPORT LIBRARIES ======================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import math
import warnings
warnings.filterwarnings("ignore")

# ====================== 2. SET VISUALIZATION STYLE ======================
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = [12, 6]
plt.rcParams['figure.dpi'] = 100

# ====================== 3. DATA LOADING AND PREPROCESSING ======================
def load_data(filepath):
    df = pd.read_csv(filepath)
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    df.sort_index(inplace=True)
    return df

# ====================== 4. LSTM DATA PREPARATION ======================
def preprocess_lstm_data(df, look_back=60, train_ratio=0.8):
    dataset = df['Close'].values.reshape(-1, 1).astype('float32')
    scaler = MinMaxScaler(feature_range=(0, 1))
    dataset = scaler.fit_transform(dataset)

    # Split into training and test sets
    train_size = int(len(dataset) * train_ratio)
    train, test = dataset[0:train_size,:], dataset[train_size:,:]

    # Create time series sequences
    def create_dataset(data, look_back=1):
        X, Y = [], []
        for i in range(len(data)-look_back-1):
            X.append(data[i:(i+look_back), 0])
            Y.append(data[i + look_back, 0])
        return np.array(X), np.array(Y)

    X_train, y_train = create_dataset(train, look_back)
    X_test, y_test = create_dataset(test, look_back)

    # Reshape for LSTM input [samples, time steps, features]
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    return X_train, y_train, X_test, y_test, scaler, train_size

# ====================== 5. MODEL ARCHITECTURES ======================

# 5.1 LSTM Model
def build_lstm_model(look_back):
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=(look_back, 1)),
        Dropout(0.3),
        LSTM(32),
        Dropout(0.3),
        Dense(1)
    ])
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# 5.2 XGBoost Model (Replaces Prophet)
def train_xgboost(train_series, test_series=None, look_back=60):
    """
    XGBoost model for time series forecasting
    Args:
        train_series: Training data (Pandas Series with datetime index)
        test_series: Optional test data for validation
        look_back: Number of previous days to use as features
    Returns:
        Predictions for the test period
    """
    # Create features using same look-back window as other models
    def create_features(data, look_back):
        X, y = [], []
        for i in range(len(data)-look_back-1):
            X.append(data.iloc[i:i+look_back].values)
            y.append(data.iloc[i+look_back])
        return np.array(X), np.array(y)

    # Prepare training data
    X_train, y_train = create_features(train_series, look_back)

    # Initialize and fit model
    model = XGBRegressor(
        n_estimators=100,
        learning_rate=0.1,
        max_depth=5,
        random_state=42
    )
    model.fit(X_train, y_train)

    # Make predictions if test data provided
    if test_series is not None:
        # Combine train and test for feature creation
        full_series = pd.concat([train_series, test_series])
        X_test, _ = create_features(full_series, look_back)
        X_test = X_test[-len(test_series):]  # Take only the test portion
        return model.predict(X_test)
    return None

# 5.3 Random Forest Model
def train_random_forest(X_train, y_train, n_estimators=100):
    model = RandomForestRegressor(n_estimators=n_estimators, random_state=42)
    model.fit(X_train, y_train)
    return model

# ====================== 6. MODEL EVALUATION ======================
def evaluate_models(true_values, predictions, model_name):
    mse = mean_squared_error(true_values, predictions)
    rmse = math.sqrt(mse)
    mae = mean_absolute_error(true_values, predictions)
    r2 = r2_score(true_values, predictions)

    print(f"\n{model_name} Performance:")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAE: {mae:.2f}")
    print(f"RÂ² Score: {r2:.2f}")

    return {'RMSE': rmse, 'MAE': mae, 'R2': r2}

# ====================== 7. VISUALIZATION FUNCTIONS ======================

# 7.1 Model Comparison Plot
def plot_comparison(df, lstm_pred, xgb_pred, rf_pred, train_size, look_back):
    # Prepare dates for plotting
    dates = df.index[look_back:train_size+look_back+len(lstm_pred)]

    # Create figure
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

    # Plot training period
    ax1.plot(df.index[:train_size], df['Close'][:train_size], label='Actual (Train)', color='blue')
    ax1.set_title('Training Period Performance')
    ax1.legend()
    ax1.grid(True)

    # Plot test period
    test_dates = df.index[train_size+look_back+1:train_size+look_back+1+len(lstm_pred)]
    ax2.plot(test_dates, df['Close'][train_size+look_back+1:train_size+look_back+1+len(lstm_pred)],
             label='Actual (Test)', color='blue')
    ax2.plot(test_dates, lstm_pred, label='LSTM Prediction', color='green')
    ax2.plot(test_dates, xgb_pred, label='XGBoost Prediction', color='red')
    ax2.plot(test_dates, rf_pred, label='Random Forest Prediction', color='purple')
    ax2.set_title('Test Period Performance Comparison')
    ax2.legend()
    ax2.grid(True)

    # Format x-axis
    for ax in [ax1, ax2]:
        ax.xaxis.set_major_locator(mdates.YearLocator())
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
        plt.setp(ax.get_xticklabels(), rotation=45)

    plt.tight_layout()
    plt.savefig('model_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

# 7.2 Metrics Comparison Plot
def plot_metrics(metrics):
    metrics_df = pd.DataFrame(metrics).T
    metrics_df.plot(kind='bar', subplots=True, layout=(1,3), figsize=(18,5))
    plt.suptitle('Model Performance Metrics Comparison')
    plt.tight_layout()
    plt.savefig('performance_metrics.png', dpi=300, bbox_inches='tight')
    plt.show()

# ====================== 8. MAIN EXECUTION ======================
def main():
    # Parameters
    look_back = 60
    train_ratio = 0.8
    epochs = 100
    batch_size = 32

    # Load and preprocess data
    df = load_data('apple_stock.csv')
    X_train_lstm, y_train_lstm, X_test_lstm, y_test_lstm, scaler, train_size = preprocess_lstm_data(df, look_back, train_ratio)

    # Prepare data for XGBoost and Random Forest
    train_data = df['Close'][:train_size]
    test_data = df['Close'][train_size+look_back+1:train_size+look_back+1+len(X_test_lstm)]

    # Prepare features for Random Forest
    def create_rf_features(data, look_back):
        X, y = [], []
        for i in range(len(data)-look_back-1):
            X.append(data.iloc[i:i+look_back].values)
            y.append(data.iloc[i+look_back])
        return np.array(X), np.array(y)

    X_train_rf, y_train_rf = create_rf_features(df['Close'][:train_size], look_back)
    X_test_rf, y_test_rf = create_rf_features(df['Close'][train_size-look_back:], look_back)

    # Train LSTM model
    print("Training LSTM model...")
    lstm_model = build_lstm_model(look_back)
    lstm_model.fit(X_train_lstm, y_train_lstm,
                  validation_data=(X_test_lstm, y_test_lstm),
                  epochs=epochs, batch_size=batch_size, verbose=1)

    # Train XGBoost model (using same look-back as other models)
    print("\nTraining XGBoost model...")
    xgb_pred = train_xgboost(train_data, test_data, look_back)

    # Train Random Forest
    print("\nTraining Random Forest model...")
    rf_model = train_random_forest(X_train_rf, y_train_rf)

    # Make predictions
    lstm_pred = lstm_model.predict(X_test_lstm)
    lstm_pred = scaler.inverse_transform(lstm_pred).flatten()
    rf_pred = rf_model.predict(X_test_rf[:len(test_data)])

    # Evaluate models
    metrics = {}
    metrics['LSTM'] = evaluate_models(test_data.values, lstm_pred, "LSTM")
    metrics['XGBoost'] = evaluate_models(test_data.values, xgb_pred, "XGBoost")
    metrics['Random Forest'] = evaluate_models(test_data.values, rf_pred, "Random Forest")

    # Generate visualizations
    plot_comparison(df, lstm_pred, xgb_pred, rf_pred, train_size, look_back)
    plot_metrics(metrics)

    return metrics

if __name__ == '__main__':
    metrics = main()
```

# Training the LSTM Model and storing it for later use

```python
# ====================== 1. IMPORT LIBRARIES ======================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import math

# ====================== 2. DATA LOADING AND PREPROCESSING ======================

def load_data(filepath):
    """
    Load and prepare the stock price dataset
    Args:
        filepath: Path to the CSV file containing stock data
    Returns:
        DataFrame with Date as index
    """
    df = pd.read_csv(filepath)
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    df.sort_index(inplace=True)
    return df

def preprocess_data(df, look_back=60, train_ratio=0.8):
    """
    Prepare data for LSTM model training
    Args:
        df: DataFrame with stock prices
        look_back: Number of previous time steps to use
        train_ratio: Ratio of training to test data
    Returns:
        Processed training and test data with scaler
    """
    # Use 'Close' price for prediction
    dataset = df['Close'].values.reshape(-1, 1)
    dataset = dataset.astype('float32')

    # Normalize the dataset
    scaler = MinMaxScaler(feature_range=(0, 1))
    dataset = scaler.fit_transform(dataset)

    # Split into train and test sets
    train_size = int(len(dataset) * train_ratio)
    test_size = len(dataset) - train_size
    train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]

    # Create dataset with look_back
    def create_dataset(dataset, look_back=1):
        X, Y = [], []
        for i in range(len(dataset)-look_back-1):
            a = dataset[i:(i+look_back), 0]
            X.append(a)
            Y.append(dataset[i + look_back, 0])
        return np.array(X), np.array(Y)

    X_train, y_train = create_dataset(train, look_back)
    X_test, y_test = create_dataset(test, look_back)

    # Reshape input to be [samples, time steps, features]
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    return X_train, y_train, X_test, y_test, scaler

# ====================== 3. MODEL ARCHITECTURE AND TRAINING ======================

def build_model(look_back):
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(look_back, 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(50, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(50))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

def train_model(model, X_train, y_train, X_test, y_test, epochs=100, batch_size=32):
    """
    Train the LSTM model with early stopping
    Args:
        model: Compiled Keras model
        X_train, y_train: Training data
        X_test, y_test: Validation data
        epochs: Maximum training epochs
        batch_size: Batch size for training
    Returns:
        Training history object
    """
    # Callbacks
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stop, checkpoint],
        verbose=1
    )
    return history

# ====================== 4. MODEL EVALUATION ======================

def evaluate_model(model, X_train, y_train, X_test, y_test, scaler):
    """
    Evaluate model performance
    Args:
        model: Trained Keras model
        X_train, y_train: Training data
        X_test, y_test: Test data
        scaler: Scaler used for normalization
    Returns:
        Predictions for training and test sets
    """
    # Make predictions
    train_predict = model.predict(X_train)
    test_predict = model.predict(X_test)

    # Invert predictions
    train_predict = scaler.inverse_transform(train_predict)
    y_train = scaler.inverse_transform([y_train])
    test_predict = scaler.inverse_transform(test_predict)
    y_test = scaler.inverse_transform([y_test])

    # Calculate RMSE
    train_score = math.sqrt(mean_squared_error(y_train[0], train_predict[:,0]))
    print(f'Train Score: {train_score:.2f} RMSE')
    test_score = math.sqrt(mean_squared_error(y_test[0], test_predict[:,0]))
    print(f'Test Score: {test_score:.2f} RMSE')

    return train_predict, test_predict

# ====================== 5. VISUALIZATION ======================

def plot_results(df, train_predict, test_predict, look_back):
    """
    Visualize actual vs predicted prices
    Args:
        df: Original DataFrame
        train_predict: Training set predictions
        test_predict: Test set predictions
        look_back: Number of time steps used
    """
    # Shift train predictions for plotting
    train_predict_plot = np.empty_like(df['Close'].values)
    train_predict_plot[:] = np.nan
    train_predict_plot[look_back:len(train_predict)+look_back] = train_predict[:,0]

    # Shift test predictions for plotting
    test_predict_plot = np.empty_like(df['Close'].values)
    test_predict_plot[:] = np.nan
    test_predict_plot[len(train_predict)+(look_back*2)+1:len(df)-1] = test_predict[:,0]

    # Plot baseline and predictions
    plt.figure(figsize=(16,8))
    plt.plot(df['Close'].values, label='Actual Price')
    plt.plot(train_predict_plot, label='Training Prediction')
    plt.plot(test_predict_plot, label='Testing Prediction')
    plt.title('Apple Stock Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

# ====================== 6. MAIN EXECUTION ======================

def main():
    # Parameters
    look_back = 60  # Number of previous days to use for prediction
    epochs = 100
    batch_size = 32

    # Load and preprocess data
    df = load_data('apple_stock.csv')
    X_train, y_train, X_test, y_test, scaler = preprocess_data(df, look_back)

    # Build model
    model = build_model(look_back)

    # Train model
    history = train_model(model, X_train, y_train, X_test, y_test, epochs, batch_size)

    # Evaluate model
    train_predict, test_predict = evaluate_model(model, X_train, y_train, X_test, y_test, scaler)

    # Plot results
    plot_results(df, train_predict, test_predict, look_back)

    return model, scaler

if __name__ == '__main__':
    model, scaler = main()
```

# Investment Strategy

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.metrics import mean_squared_error
from ta import add_all_ta_features
import warnings
warnings.filterwarnings("ignore")

# ====================== 1. Data Loading & Preprocessing ======================
def load_and_preprocess_data(filepath):
    df = pd.read_csv(filepath, parse_dates=["Date"], index_col="Date")
    df.sort_index(inplace=True)

    if 'Volume' not in df.columns:
        df['Volume'] = 0

    df = add_all_ta_features(
        df, open="Open", high="High", low="Low",
        close="Close", volume="Volume", fillna=True
    )

    features = ["Close", "momentum_rsi", "trend_macd", "volatility_bbh", "volatility_bbl"]
    return df[[f for f in features if f in df.columns]].dropna()

# ====================== 2. Enhanced Trading Strategy ======================
class EnhancedTradingStrategy:
    def __init__(self, test_df, predictions):
        self.df = test_df.copy()
        self.predictions = predictions
        self.signals = pd.DataFrame(index=test_df.index)
        self.signals["price"] = test_df["Close"]
        self.signals["predicted"] = predictions
        self.signals["signal"] = 0

    def generate_signals(self):
        # Calculate indicators
        self.signals["pred_diff"] = self.signals["predicted"].pct_change(3)
        self.signals["price_diff"] = self.signals["price"].pct_change(3)

        # Buy conditions
        buy_condition = (
            (self.signals["pred_diff"] > 0.05) &
            (self.signals["price_diff"] > 0) &
            (self.df["momentum_rsi"] < 45)
        )

        # Sell conditions
        sell_condition = (
            (self.signals["pred_diff"] < -0.03) &
            (self.signals["price_diff"] < 0) &
            (self.df["momentum_rsi"] > 60)
        )

        self.signals["signal"] = np.select(
            [buy_condition, sell_condition],
            [1, -1],
            default=0
        )

        self.signals["position"] = self.signals["signal"].replace(0, np.nan).ffill().fillna(0)
        return self.signals

    def calculate_returns(self):
        self.signals["market_return"] = self.signals["price"].pct_change()
        self.signals["strategy_return"] = (
            self.signals["position"].shift(1) * self.signals["market_return"] -
            self.signals["position"].diff().abs() * 0.001
        )

        self.signals["portfolio_value"] = 10000 * (1 + self.signals["strategy_return"]).cumprod()
        self.signals["buy_hold_value"] = 10000 * (1 + self.signals["market_return"]).cumprod()
        return self.signals

# ====================== 3. Visualization Functions ======================
def plot_predictions_vs_actual(df, predictions, split_date):
    plt.figure(figsize=(14, 7))
    plt.plot(df.index, df["Close"], label="Actual Price", alpha=0.7)
    plt.plot(predictions.index, predictions, label="Predicted Price", alpha=0.7)
    plt.axvline(split_date, color="gray", linestyle="--", label="Train/Test Split")
    plt.title("Actual vs Predicted Prices")
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_test_signals(signals):
    plt.figure(figsize=(14, 7))
    plt.plot(signals.index, signals["price"], label="Price", alpha=0.7)

    buy_signals = signals[signals["signal"] > 0]
    sell_signals = signals[signals["signal"] < 0]

    plt.scatter(buy_signals.index, buy_signals["price"],
               marker="^", color="green", s=100, label="Buy")
    plt.scatter(sell_signals.index, sell_signals["price"],
               marker="v", color="red", s=100, label="Sell")

    plt.title("Trading Signals")
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_test_performance(signals):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

    ax1.plot(signals.index, signals["portfolio_value"], label="Strategy")
    ax1.plot(signals.index, signals["buy_hold_value"], label="Buy & Hold")
    ax1.set_title("Performance Comparison")
    ax1.legend()
    ax1.grid(True)

    cumulative_max = signals["portfolio_value"].cummax()
    drawdown = (signals["portfolio_value"] - cumulative_max) / cumulative_max
    ax2.fill_between(signals.index, drawdown * 100, 0, color="red", alpha=0.3)
    ax2.set_title("Strategy Drawdown (%)")
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

# ====================== 4. Main Execution ======================
if __name__ == "__main__":
    # Load and prepare data
    df = load_and_preprocess_data("apple_stock.csv")

    # Normalize and create sequences
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df)
    look_back = 60
    X, y = [], []
    for i in range(look_back, len(scaled_data)):
        X.append(scaled_data[i-look_back:i])
        y.append(scaled_data[i, 0])
    X, y = np.array(X), np.array(y)

    # Train-test split
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]
    test_start_idx = look_back + split
    test_df = df.iloc[test_start_idx:]

    # Build and train model
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=(look_back, X.shape[2])),
        Dropout(0.3),
        LSTM(32),
        Dropout(0.3),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")
    model.fit(X_train, y_train, epochs=50, batch_size=32,
              validation_data=(X_test, y_test), verbose=1)

    # Generate predictions for test period only
    test_predictions = model.predict(X_test)
    test_predictions = scaler.inverse_transform(
        np.concatenate((test_predictions, np.zeros((len(test_predictions), len(df.columns)-1))), axis=1))[:, 0]
    test_predictions = pd.Series(test_predictions, index=test_df.index)

    # Run strategy on test period
    strategy = EnhancedTradingStrategy(test_df, test_predictions)
    signals = strategy.generate_signals()
    returns = strategy.calculate_returns()

    # Generate visualizations
    plot_predictions_vs_actual(df.iloc[look_back:],
                             pd.concat([pd.Series([np.nan]*split, index=df.index[look_back:test_start_idx]),
                                        test_predictions]),
                             test_df.index[0])
    plot_test_signals(signals)
    plot_test_performance(returns)

    # Performance metrics
    sharpe = returns["strategy_return"].mean() / returns["strategy_return"].std() * np.sqrt(252)
    bh_return = (returns["buy_hold_value"].iloc[-1] / 10000) ** (252/len(returns)) - 1
    strat_return = (returns["portfolio_value"].iloc[-1] / 10000) ** (252/len(returns)) - 1

    print(f"\n=== Performance Metrics ===")
    print(f"Strategy Annualized Return: {strat_return:.2%}")
    print(f"Buy & Hold Annualized Return: {bh_return:.2%}")
    print(f"Strategy Sharpe Ratio: {sharpe:.2f}")
```

## Implementation Notes

### Why Functions Were Used:
- **Consistency**: Ensures identical preprocessing for all models  
- **Comparability**: Enables fair evaluation across different approaches  
- **Extensibility**: Easy to add new models to the comparison framework  
- **Reproducibility**: Self-contained units for each modeling technique  

### Optimization Opportunities:
- **Feature Enhancement**: Add volume and technical indicators  
- **Hybrid Approach**: Combine LSTM predictions with XGBoost residuals  
- **Hyperparameter Tuning**: Bayesian optimization for all models  
- **Trading Simulation**: Implement backtesting with transaction costs  

> **Note**: While LSTM showed superior performance in our tests, all models could be enhanced with additional feature engineering and parameter tuning. The functional implementation allows for systematic improvement of each approach while maintaining comparability.
